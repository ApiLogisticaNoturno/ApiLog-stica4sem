```python
# CRIANDO O CSV 2023 + 2024 - ATRAC
arquivos = [file1, file2]
ANTAQ_Int = pd.DataFrame()

for arquivo in arquivos:
    #AJUSTANDO ARQUIVO
    df = pd.read_csv(arquivo, sep=';', encoding='UTF-8', decimal=',')
    ANTAQ_Int = pd.concat([ANTAQ_Int, df], ignore_index=True)

#VISUALIZAR ARQUIVO
print(ANTAQ_Int.head(5))
print(ANTAQ_Int.tail(5))
```

```python
#SELECIONAR AS COLUNAS DE INTERESSE

colunas_interesse = ['IDAtracacao','IDBerco',
                     'Data Atracação', 'Data Chegada',
                     'Data Desatracação','Data Início Operação',
                     'Data Término Operação']

ANTAQ = ANTAQ_Int.dropna(subset=colunas_interesse)
ANTAQ.info()
```

```python
#CONFIGURAR AS COLUNAS DE DATAS/HORAS
ANTAQ['Data Chegada'] = pd.to_datetime(ANTAQ['Data Chegada'], format='%d/%m/%Y %H:%M:%S')
ANTAQ['Data Atracação'] = pd.to_datetime(ANTAQ['Data Atracação'], format='%d/%m/%Y %H:%M:%S')
ANTAQ['Data Início Operação'] = pd.to_datetime(ANTAQ['Data Início Operação'], format='%d/%m/%Y %H:%M:%S')
ANTAQ['Data Término Operação'] = pd.to_datetime(ANTAQ['Data Término Operação'], format='%d/%m/%Y %H:%M:%S')
ANTAQ['Data Desatracação'] = pd.to_datetime(ANTAQ['Data Desatracação'], format='%d/%m/%Y %H:%M:%S')
```

```python
#CRIANDO COLUNAS DE INTERESSE DATAS/HORAS
ANTAQ['Tempo_espera_chegada'] = (ANTAQ['Data Atracação'] - ANTAQ['Data Chegada']).dt.total_seconds() / 60
ANTAQ['Tempo_espera_inicial'] = (ANTAQ['Data Início Operação'] - ANTAQ['Data Atracação']).dt.total_seconds() / 60
ANTAQ['Tempo_operacao'] = (ANTAQ['Data Término Operação'] - ANTAQ['Data Início Operação']).dt.total_seconds() / 60
ANTAQ['Tempo_espera_final'] = (ANTAQ['Data Desatracação'] - ANTAQ['Data Término Operação']).dt.total_seconds() / 60
ANTAQ['Tempo_ocupacao'] = (ANTAQ['Data Desatracação'] - ANTAQ['Data Atracação']).dt.total_seconds() / 60
```

```python
#INSERIR TABELA SH4 DA NOMENCLATURAS DAS MERCADORIAS
Mercadorias = caminho_in+'Mercadorias.csv'
with open(Mercadorias, 'r', encoding='UTF-8') as f:
    linhas = f.readlines()
    for i in range(0, 5):
        print(f"Linha {i+1}: {linhas[i]}")

Mercadorias = pd.read_csv(Mercadorias, sep=';', encoding='UTF-8', decimal=',')
```

```python
#REALIZAR JOIN DA TABELA CARGAS COM A TABELA MERCADORIAS
#MANTER SOMENTE A COLUNA MERCADORIA
mercadorias_reduzida = Mercadorias[['CDMercadoria', 'Mercadoria']]

#REALIZAR A JUNÇÃO DAS TABELAS
df_merged = pd.merge(Carga_Int, mercadorias_reduzida, on='CDMercadoria', how='left')

#CONFERIR
print(df_merged.head(10))
```

```python
#AGRUPAR OS VALORES DO DATA FRAME DE ACORDO COM ATRACAÇÃO E TIPO DE PRODUTO (SOJA)
# CARREGAR TODOS OS DADOS
df5 = df_merged.copy()

# CONVERTER VALORES PARA FLOAT
df5['VLPesoCargaBruta'] = df5['VLPesoCargaBruta'].astype(str) \
    .str.replace('.', '', regex=False) \
    .str.replace(',', '.', regex=False) \
    .astype(float)


# CARREGAR DATA FRAME PARA SOJA
df_soja = df5[df5['CDMercadoria'] == '1201']

# CARREGAR DATA FRAME PARA TODAS AS MERCADORIAS
df_todas_mercadorias = df5

#  AGRUPAMENTO DO PESO NO DATA FRAME SOJA
colunas = ['IDAtracacao', 'CDMercadoria']
coluna_valor = 'VLPesoCargaBruta'
cargas_soja = df_soja.groupby(colunas)[coluna_valor].sum().reset_index()

# TODAS AS MERCADORIAS SEM FILTRO:
cargas_todas = df_todas_mercadorias.groupby(colunas)[coluna_valor].sum().reset_index()

# ORDENAÇAO DOS DATA FRAMES
cargas_soja = cargas_soja.sort_values(by=coluna_valor, ascending=False)
cargas_todas = cargas_todas.sort_values(by=coluna_valor, ascending=False)
```

```python
# Junção entre ANTAQ e df_soja com base na coluna 'IDAtracacao' (inner join)
def merge_e_filtrar(df_base, df_merge, colunas_chave, colunas_desejadas, nome="resultado"):
    """
    Realiza merge entre dois DataFrames e filtra apenas as colunas desejadas.
    Reduz uso de memória selecionando previamente as colunas envolvidas.

    Parâmetros:
    - df_base: DataFrame principal (ex: ANTAQ)
    - df_merge: DataFrame auxiliar (ex: df_soja ou df_todas_mercadorias)
    - colunas_chave: coluna(s) para merge (ex: 'IDAtracacao')
    - colunas_desejadas: lista final de colunas desejadas
    - nome: nome para mensagens de depuração

    Retorna:
    - DataFrame final filtrado
    """

    # Garante que colunas_chave seja lista
    if isinstance(colunas_chave, str):
        colunas_chave = [colunas_chave]

    # Selecionar apenas colunas chave + desejadas de cada DataFrame
    colunas_base = list(set(colunas_chave + colunas_desejadas) & set(df_base.columns))
    colunas_merge = list(set(colunas_chave + colunas_desejadas) & set(df_merge.columns))

    df_base_reduzido = df_base[colunas_base].copy()
    df_merge_reduzido = df_merge[colunas_merge].copy()

    # Realiza merge com base nas colunas chave
    try:
        df_final = pd.merge(df_base_reduzido, df_merge_reduzido, on=colunas_chave, how='inner')
    except Exception as e:
        print(f"❌ Erro ao realizar merge no DataFrame '{nome}': {e}")
        return pd.DataFrame()

    # Filtra as colunas desejadas (se existirem)
    colunas_disponiveis = [col for col in colunas_desejadas if col in df_final.columns]
    df_final = df_final[colunas_disponiveis]

    # Aviso sobre colunas ausentes
    colunas_faltando = list(set(colunas_desejadas) - set(df_final.columns))
    if colunas_faltando:
        print(f"⚠️ Atenção: As seguintes colunas estão ausentes no DataFrame '{nome}' após o merge:")
        for col in colunas_faltando:
            print(f"  - {col}")

    return df_final

# Lista das colunas que você quer manter
colunas_para_manter = [
    'IDAtracacao', 'IDBerco', 'Berço', 'Porto Atracação', 'Coordenadas', 'Complexo Portuário',
    'Data Atracação', 'Data Chegada', 'Data Desatracação', 'Data Início Operação', 'Data Término Operação',
    'Ano', 'Mes', 'Tipo de Operação', 'Terminal', 'Município', 'UF', 'SGUF', 'Região Geográfica',
    'Região Hidrográfica', 'Tempo_espera_chegada', 'Tempo_espera_inicial', 'Tempo_operacao',
    'Tempo_espera_final', 'Tempo_ocupacao', 'IDCarga', 'Origem', 'Destino',
    'CDMercadoria', 'Tipo Operação da Carga', 'Natureza da Carga', 'VLPesoCargaBruta', 'Mercadoria'
]

# Merge e filtragem para o dataframe de soja
final_soja = merge_e_filtrar(ANTAQ, df_soja, colunas_chave='IDAtracacao',
                              colunas_desejadas=colunas_para_manter, nome='final_soja')

# Merge e filtragem para o dataframe de todas as mercadorias
final_todas = merge_e_filtrar(ANTAQ, df_todas_mercadorias, colunas_chave='IDAtracacao',
                              colunas_desejadas=colunas_para_manter, nome='final_todas')


```
